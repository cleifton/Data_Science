```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
# Obter os dados do IBOV
ibov = yf.download('^BVSP', start='2021-01-01', end='2024-03-18')
```

```python
# Calcular médias móveis
ibov['SMA_21'] = ibov['Close'].rolling(window=21).mean()
ibov['SMA_80'] = ibov['Close'].rolling(window=80).mean()
```

```python
# Calcular RSI (Índice de Força Relativa)
delta = ibov['Close'].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
rs = gain / loss
ibov['RSI'] = 100 - (100 / (1 + rs))
```

```python
# Calcular MACD (Divergência de Convergência da Média Móvel)
shortEMA = ibov['Close'].ewm(span=12, adjust=False).mean()
longEMA = ibov['Close'].ewm(span=26, adjust=False).mean()
ibov['MACD'] = shortEMA - longEMA
ibov['Signal Line'] = ibov['MACD'].ewm(span=9, adjust=False).mean()
```

```python
# Construir modelo preditivo simples
ibov.loc[ibov['SMA_21'] > ibov['SMA_80'], 'Signal'] = 1  # Comprar quando a média de 21 é maior que a de 80
ibov.loc[ibov['SMA_21'] < ibov['SMA_80'], 'Signal'] = -1  # Vender quando a média de 21 é menor que a de 80
```

```python
# Plotar os indicadores e os sinais de compra/venda
plt.figure(figsize=(16, 8))
plt.plot(ibov['Close'], label='Close Price')
plt.plot(ibov['SMA_21'], label='SMA 21')
plt.plot(ibov['SMA_80'], label='SMA 80')
```

```python
# Adicionar pontos de compra e venda ao gráfico
buy_signals = ibov[ibov['Signal'] == 1]
sell_signals = ibov[ibov['Signal'] == -1]

plt.scatter(buy_signals.index, buy_signals['Close'], marker='^', color='g', label='Buy Signal')
plt.scatter(sell_signals.index, sell_signals['Close'], marker='v', color='r', label='Sell Signal')

plt.legend()
plt.show()

plt.figure(figsize=(16, 4))
plt.plot(ibov['RSI'], label='RSI')
```

```python
# Adicionar linhas de sobrecompra e sobrevenda ao gráfico
plt.axhline(y=70, color='r', linestyle='--', label='Overbought')
plt.axhline(y=30, color='g', linestyle='--', label='Oversold')

plt.legend()
plt.show()

plt.figure(figsize=(16, 4))
plt.plot(ibov['MACD'], label='MACD')
plt.plot(ibov['Signal Line'], label='Signal Line')
```

```python
# Adicionar linha zero ao gráfico
plt.axhline(y=0, color='gray', linestyle='--', label='Zero')

plt.legend()
plt.show()
```
