```python
#Analise de Petr4

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as web

!pip install yfinance --upgrade --no-cache-dir
import yfinance as yf
yf.pdr_override()
```

```python
ibov = web.get_data_yahoo('PETR4.SA', start='2019-05-03', end='2022-10-20')
```

```python
ibov.head()
```

```python
ibov.tail()
```

```python
ibov["Close"].plot(figsize=(22,8), label="PETR4")
ibov["Close"].rolling(21).mean().plot(label="MM21")
ibov["Close"].rolling(80).mean().plot(label="MM80")
plt.legend();
```

```python
ibov_fatiado = ibov[ibov.index.year == 2021]
ibov_fatiado["Close"].plot(figsize=(22,8), label="PETR4")
ibov_fatiado["Close"].rolling(21).mean().plot(label="MM21")
ibov_fatiado["Close"].rolling(80).mean().plot(label="MM80")
plt.legend();
```

```python
ibov_fatiado = ibov[(ibov.index.year >= 2019) & (ibov.index.year <= 2023)]
ibov_fatiado["Close"].plot(figsize=(22,8), label="PETR4")
ibov_fatiado["Close"].rolling(21).mean().plot(label="MM21")
ibov_fatiado["Close"].rolling(80).mean().plot(label="MM80")
plt.legend();
```

```python
ibov_fatiado = ibov[ibov.index.month == 12]
ibov_fatiado.tail(20)
```

```python
ibov_fatiado = ibov[ibov.index.day == 15]
ibov_fatiado.tail(30)
```
